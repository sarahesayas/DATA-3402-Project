```python
import matplotlib.pyplot as plt

def compare_pawpularity(feature_name="Eyes"):
    plt.hist(data["Pawpularity"][data[feature_name]==1], bins=50, density=1., label=feature_name+"==1", alpha=0.5)
    plt.hist(data["Pawpularity"][data[feature_name]!=1], bins=50, density=1., label=feature_name+"!=1", alpha=0.5)

# Set up a grid for subplots
fig, axs = plt.subplots(3, 4, figsize=(15, 10))

# Flatten the axs array to iterate over all subplots
axs = axs.flatten()

# Plot each histogram in a separate subplot
for i, feature_name in enumerate(data.columns[2:-1]):
    axs[i].hist(data["Pawpularity"][data[feature_name]==1], bins=50, density=1., label=feature_name+"==1", alpha=0.5)
    axs[i].hist(data["Pawpularity"][data[feature_name]!=1], bins=50, density=1., label=feature_name+"!=1", alpha=0.5)
    axs[i].set_title(f'Histogram for {feature_name}')
    axs[i].legend()

# Hide the extra subplots in the last row if they're not needed
for j in range(len(data.columns[2:-1]), len(axs)):
    axs[j].axis('off')

plt.tight_layout()
plt.show()
```
